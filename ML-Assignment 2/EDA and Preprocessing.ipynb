## Data Cleaning
### Find the missing and inappropriate values, treat them appropriately.


```python
## dropping duplicates 
df.drop_duplicates(inplace =True)
df
```

```python
## Detecting Outliers 

numerical_columns=["Age","Salary","Gender"]
plt.figure(figsize=(8,6))
sns.boxplot(data=df[numerical_columns])
plt.title ("Outlier Detection")
plt.show()
```

## Replace the value 0 in age as NaN


```python

df['Age'] = df['Age'].replace(0, np.nan)
df
```

## Treat the null values in all columns using any measures(removing/ replace the values with mean/median/mode)

```python
# filling the missing values with mean of each column

for col in ["Age","Salary"]:
    df[col]= df[col].fillna(df[col].mean())

df.head(35)
```

```python
# filling the missing values with mode of each  categorical column

for col in ['Company','Place']:
    df[col]=df[col].fillna(df[col].mode()[0] )

df.head(36)
```

```python
df.isnull().sum()
```

## Data Encoding:Convert categorical variables into numerical representations using techniques such as one-hot encoding, label encoding, making them suitable for analysis by machine learning algorithms.


```python
from sklearn.preprocessing  import OneHotEncoder,LabelEncoder 
```

```python
##initialising One Hot Encoder 

ohe = OneHotEncoder(sparse_output =False)
```

```python
# Fit and transform the data 

encoded= ohe.fit_transform (df[['Company','Place']])

encoded_df = pd.DataFrame(encoded,columns=ohe.get_feature_names_out(ohe.feature_names_in_ ))

encoded_df
```

```python
## Adding new DataFrame to the Old Numeric Data

numeric_data = df[['Age','Salary']]

new_df = pd.concat([numeric_data,encoded_df] ,axis = 1)

new_df
```

## Feature Scaling:
After the process of encoding, perform the scaling of the features using standardscaler and minmaxscaler.

```python
  from sklearn.preprocessing import StandardScaler,MinMaxScaler
```
